```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# =============================
# 1. data
# =============================
rows = ["C6", "C5", "C4", "C3", "C2", "C1"]
values = [0.8764, 0.8694, 0.8664, 0.8492, 0.8434, 0.6864]
groups = ["PDAC", "duct", "pancreas", "demo"]

# 1 = drop, draw X
# 0 = keep, draw empty square
drop_matrix = [
    [0, 0, 0, 0],  # C6
    [0, 0, 1, 0],  # C5
    [0, 0, 0, 1],  # C4
    [0, 1, 0, 0],  # C3
    [1, 0, 0, 0],  # C2
    [0, 1, 1, 1],  # C1
]

# =============================
# 2. style
# =============================
sns.set_theme(style="white")
plt.rcParams["font.family"] = "DejaVu Sans"

edge_color = "#4F7A94"
value_color = "#587283"

bar_colors = [
    "#ff8802",  # C6
    "#EBEFF1",  # C5
    "#EBEFF1",  # C4
    "#EBEFF1",  # C3
    "#EBEFF1",  # C2
    "#BFC5CD",  # C1
]

# =============================
# 3. layout
# =============================
y_step = 0.90
y = np.arange(len(rows)) * y_step

fig = plt.figure(figsize=(9.2, 4.8), dpi=140)
gs = fig.add_gridspec(1, 2, width_ratios=[1.20, 5.10], wspace=0.01)

ax_left = fig.add_subplot(gs[0, 0])
ax_bar = fig.add_subplot(gs[0, 1], sharey=ax_left)

# =============================
# 4. right: horizontal bars
# =============================
x_min = 0.682
x_max = 0.884
widths = np.array(values) - x_min

ax_bar.barh(
    y,
    widths,
    left=x_min,
    height=0.52,
    color=bar_colors,
    edgecolor="#C5CCD2",
    linewidth=1.0
)

ax_bar.set_xlim(x_min, x_max)

for yi, val in zip(y, values):
    dx = 0.002 if val < 0.75 else 0.0025
    ax_bar.text(
        val + dx,
        yi,
        f"{val:.4f}",
        va="center",
        ha="left",
        fontsize=12.5,
        color=value_color,
        fontweight="bold"
    )

ax_bar.set_xlabel("AUROC", fontsize=14, fontweight="bold")
ax_bar.set_yticks(y)
ax_bar.set_yticklabels([])
ax_bar.tick_params(axis="y", length=0)

ax_bar.spines["top"].set_visible(False)
ax_bar.spines["right"].set_visible(False)
ax_bar.spines["left"].set_linewidth(1.4)
ax_bar.spines["bottom"].set_linewidth(1.4)

ticks = [0.682, 0.750, 0.817, 0.884]
ax_bar.set_xticks(ticks)
ax_bar.set_xticklabels([f"{t:.3f}" for t in ticks], fontsize=10)

# =============================
# 5. left: labels + square boxes
# =============================
# 更紧凑的 box 横向位置
box_x = [0.00, 0.16, 0.32, 0.48]
label_x = -0.16

ax_left.set_xlim(-0.24, 0.62)
ax_left.set_ylim(-0.65, y[-1] + 0.50)

# C6, C5... 标签
for yi, row_label in zip(y, rows):
    ax_left.text(
        label_x,
        yi,
        row_label,
        va="center",
        ha="center",
        fontsize=14.5,
        fontweight="bold"
    )

# 用 scatter 画空心正方形，保证视觉上真的是 square
square_size = 120   # 点面积，按需要可调小到 100 或调大到 140
x_size = 95

for yi, pattern in zip(y, drop_matrix):
    # 空心方块
    ax_left.scatter(
        box_x,
        [yi] * len(box_x),
        marker="s",
        s=square_size,
        facecolors="none",
        edgecolors=edge_color,
        linewidths=1.1,
        zorder=3
    )

    # 需要 drop 的位置画 X
    for xi, dropped in zip(box_x, pattern):
        if dropped == 1:
            ax_left.scatter(
                xi,
                yi,
                marker="x",
                s=x_size,
                c=edge_color,
                linewidths=1.2,
                zorder=4
            )

# 斜着的 group labels
header_y = -0.44
header_dx = [-0.03, -0.01, -0.005, -0.01]

for xi, label, dx in zip(box_x, groups, header_dx):
    ax_left.text(
        xi + dx,
        header_y,
        label,
        rotation=45,
        ha="left",
        va="bottom",
        fontsize=11.5,
        fontweight="bold"
    )

ax_left.axis("off")

# 第一行放最上面
ax_bar.invert_yaxis()

fig.subplots_adjust(left=0.055, right=0.985, top=0.94, bottom=0.18)

plt.show()

# 保存
plt.savefig("ablation_like_plot_4cols_refined.png", dpi=300, bbox_inches="tight")
```
